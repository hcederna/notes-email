```python
import glob


# store formatted kindle notes in this directory
KINDLE_FORMATTED_NOTES_DIRECTORY = "data/kindle/"

# store formatted learning notes in this directory
APPLE_NOTES_FORMATTED_NOTES_DIRECTORY = "data/apple_notes"
```

```python
# grab list of formatted kindle notes files
kindle_notes_paths = glob.glob(KINDLE_FORMATTED_NOTES_DIRECTORY+'/*.csv')
kindle_notes_paths
```

```python
# grab list of formatted apple notes files
apple_notes_paths = glob.glob(APPLE_NOTES_FORMATTED_NOTES_DIRECTORY+'/*.csv')
apple_notes_paths
```

```python
# create list of formatted kindle and apple notes
notes_paths = kindle_notes_paths + apple_notes_paths
notes_paths
```

```python
import pandas as pd


# concatenate kindle and apple notes csv files into one df
notes_dfs = [pd.read_csv(notes_path) for notes_path in notes_paths]
for notes_df in notes_dfs:
    display(notes_df.sample(5))
```

```python
notes = pd.concat(notes_dfs, sort=False)
notes.sample(10, random_state=123)
```

```python
# format email template
for idx, row in notes.sample(3, random_state=123).iterrows():
    print("=" * 117)
    print()
    print(row['Annotation'])
    print()
    print("-- " * 39)
    print(f"{row['Title']} | {row['Author']} | Annotated on {row['Date']}")
```

```python
import textwrap


WIDTH = 101


# adjust width of text in email template
for idx, row in notes.sample(3, random_state=123).iterrows():
    print("=" * WIDTH)
    print()
    print(textwrap.fill(row['Annotation'], width=WIDTH))
    print()
    print("-- " * round(WIDTH/3))
    print(textwrap.fill(f"{row['Title']} | {row['Author']} | Annotated on {row['Date']}", width=WIDTH))
```
