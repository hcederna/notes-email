```python
html_file = open('iphone_notes.html', 'r', encoding="ISO-8859-1")

for idx, line in enumerate(html_file):
    line = re.sub('<[^<]+?>', '',line).strip()
    if line:
        # collect data for each line --> author, title, annotation, and date 
        if "==========" in line:
            div_idx = idx
            author = None
            title = None
        elif idx == div_idx + 1:
            author = line
        elif idx == div_idx + 2:
            title = line
        else:
            try:
                date = re.search("(\d+/\d+/\d+)", line).group(1)
                line = line.strip("(" + date + ")").strip()
                print(','.join([author, title, line, date]))
            except:
                print(','.join([author, title, line]))
            print()
```

```python
html_file = open('iphone_notes.html', 'r', encoding="ISO-8859-1")
csv_file = open('iphone_notes.csv', 'w')

# add header to csv file
csv_file.write("Author,Title,Annotation,Date\n")


for idx, line in enumerate(html_file):
    # remove html tags and extraneous whitespace chars from line
    line = re.sub('<[^<]+?>', '',line).strip()
    # skip empty lines
    if line:
        # reset author and title after divider
        if "==========" in line:
            div_idx = idx
            author = None
            title = None
        # store author and title data
        # "" to ensure commas in data ok with csv format
        elif idx == div_idx + 1:
            author = '"' + line + '"'
        elif idx == div_idx + 2:
            title = '"' + line + '"'
        # write author, title, annotation, and data data to csv file
        else:
            try:
                # extract date
                date = re.search("(\d+/\d+/\d+)", line).group(1)
                line = line.strip("(" + date + ")").strip()
                csv_file.write(','.join([author, title, '"' + line + '"', '"' + date + '"']))
            except:
                csv_file.write(','.join([author, title, '"' + line + '"']))
            csv_file.write('\n')
csv_file.close()
html_file.close()
```
